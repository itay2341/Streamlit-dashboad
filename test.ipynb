```python
def agg_percent(x):
    res = []
    for i in x.tolist():
        res.append(i/x.sum())
    return res

agg_percent(pd.Series([2,2,4,4,8]))
```

```python
# percent of data sent to feed by each source each day
# fields:
# feed_id | date | src | count
df_feed_break = df_status.groupby(['feed_id', 'Date']).agg({
    'src': list,
    'count': [agg_percent, np.sum]
})
df = df_feed_break.reset_index()
df = df.reset_index(drop=True)
display(df[df['Date'] == '2023-07-14'].shape,
df[df['Date'] == '2023-07-18'])
```

```python
df_us = df_status.groupby(['feed_id','Date']).agg({
    'src': list,
    'count': agg_percent
})
df_them = df_feed.groupby(['feed_id','Date']).sum()
df_feed_price = pd.merge(df_us, df_them, on=['feed_id', 'Date'], how='outer', suffixes=('_us', '_them'))

df_feed_price
```

```python
feed_price_index = df_feed_price.index.to_list()

def agg_feed_to_revenue(x):
    src, date = x.name[0], x.name[1]
    res = []
    for fi in x['feed_id']:
        if (fi, date) not in feed_price_index:
            continue
        percent_location = df_feed_price.loc[fi, date]['src'].index(src)
        percent = df_feed_price.loc[fi, date]['count'][percent_location]
        try:
            sr = df_feed_price.loc[fi, date].drop(
                ['src', 'count', 'adv_feed_id', 'advertiser']) * percent
            res.append(sr)
        except Exception as e:
            print(fi, date)
            print(df_feed_price.loc[fi, date].drop(
                ['src', 'count', 'adv_feed_id', 'advertiser']))
            print(e.args)
    if len(res) == 0:
        return np.nan
    sr = pd.DataFrame(res).sum()
    sr.drop(['CTR', 'RPM', 'Coverage'], inplace=True)
    sr.loc['CTR'] = sr['Clicks']/sr['Monetized Searches'] * 100
    sr.loc['RPM'] = sr['Revenue']/sr['Monetized Searches'] * 1000
    sr.loc['Coverage'] = sr['Monetized Searches']/sr['Total Searches'] * 100
    return sr
```

```python
df_reports = df_status.groupby(['src', 'Date']).agg({
    'count': np.sum,
    'feed_id': list,

}).apply(agg_feed_to_revenue, axis=1).dropna().apply(pd.Series).reset_index()
# df_report1.to_csv('report1.csv')
df_reports
```

```python
def save_reports(df_reports, pub_map_path):
    pubs = json.load(open(pub_map_path))
    [print(pub) for pub in pubs]
    print(df_reports.src.value_counts())
    for pub in pubs:
        dfs = []
        for lnk in pub['links']:
            print(lnk)
            print(df_reports[df_reports['src'] == lnk].shape)
            dfs.append(df_reports[df_reports['src'] == lnk])
        pd.concat(dfs).to_csv(f'reports/{pub["uuid"]}.csv', index=False)

save_reports(df_reports, 'publisher_map.json')
```
